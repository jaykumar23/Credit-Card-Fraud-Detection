Importing required libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
# loading dataset 
# ccd - credit card data
ccd = pd.read_csv("/content/creditcard.csv")
```

```python
# first 5 rows
ccd.head()
```

```python
# last 5 rows
ccd.tail()
```

```python
# checking missing values
ccd.info()
```

```python
# 0 - Normal transaction
# 1 - Fraud transaction
ccd['Class'].value_counts()
```

```python
# nt - Normal transaction
# ft - fraud transaction
nt = ccd[ccd.Class == 0]
ft = ccd[ccd.Class == 1]
print(nt.shape)
print(ft.shape)
```

```python
# Statistical measures of data
nt.Amount.describe()

```

```python
ft.Amount.describe()
```

```python
# campare nt and ft transactions
# analysing
ccd.groupby('Class').mean()
```

Building a sample dataset of nt and ft to simplify.

For this we need to take unform inputs (same size)

```python
# ntsp - normal transaction sample
# random sampling
ntsp = nt.sample(n=222)
# concatonating two dataframes
new_dataset = pd.concat([ntsp,ft],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Spliting dataset into features and targets

```python
A = new_dataset.drop(columns='Class',axis=1)
```

```python
B = new_dataset['Class']
```

```python
print(A)
print(B)
```

Spliting the data into training data and testing data

```python
A_train, A_test, B_train, B_test = train_test_split(A, B, test_size=0.2, stratify=B, random_state=2)
```

```python
print(A.shape, A_train.shape, A_test.shape)
```

```python
print(B.shape, B_train.shape, B_test.shape)
```

Model Training 


**Logistic Regression**

```python
model = LogisticRegression()
```

```python
# training model
model.fit(A_train, B_train)
```

Model Evaluation

Accuracy

```python
A_train_predict = model.predict(A_train)
training_data_accuracy = accuracy_score(A_train_predict,B_train)
```

```python
print(training_data_accuracy)
```

```python
A_test_predict = model.predict(A_test)
testing_data_accuracy = accuracy_score(A_test_predict,B_test)
```

```python
print(testing_data_accuracy)
```
